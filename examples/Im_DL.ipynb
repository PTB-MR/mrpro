```python
import pickle

# Load the lists from the files
with open('reconstructed_images_cnn.pkl', 'rb') as f:
    loaded_list_xreco_cnn = pickle.load(f)

with open('undersampled_images_cnn.pkl', 'rb') as f:
    loaded_list_xu_cnn = pickle.load(f)

with open('full_sampled_images_cnn.pkl', 'rb') as f:
    loaded_list_xf_cnn = pickle.load(f)

print('Lists have been loaded successfully.')
```

```python
import matplotlib.pyplot as plt
import torch
```

```python
# Function to plot images
def plot_images(images_reco, images_us, images_fs, indice):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(torch.abs(images_reco[indice]).detach().numpy(), cmap='gray')
    plt.title(f'Reconstructed - Image nr {indice+1}')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(torch.abs(images_us[indice]).detach().numpy(), cmap='gray')
    plt.title(f'Undersampled - Image nr {indice+1}')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(torch.abs(images_fs[indice]).detach().numpy(), cmap='gray')
    plt.title(f'Fully sampled - Image nr {indice+1}')
    plt.axis('off')
    
    plt.show()

# Plot the first and last elements of each list
plot_images(loaded_list_xreco_cnn, loaded_list_xu_cnn, loaded_list_xf_cnn, 9)
```

```python
plot_images(loaded_list_xreco_unet, loaded_list_xu_unet, loaded_list_xf_unet, 9)
```

```python
plot_images(loaded_list_xreco, loaded_list_xu, loaded_list_xf, 9)
```
